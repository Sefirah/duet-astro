# Notebook to show the band inputs from John Hennesey

```python
import astropy.units as ur
import astropy.constants as cr

import matplotlib.pyplot as plt
%matplotlib inline

from tdsat_telescope import load_qe, load_reflectivity
```

```python
for band in [1, 2, 3]:
    wave, qe= load_qe(band=band, diag = True)
    aa =  [(i.to(ur.AA)).value for i in wave]
    plt.scatter(aa, qe, label = 'Band {}'.format(band))
    plt.xlabel('Wavelength (Ang)')
    plt.ylabel('QE')


plt.legend()
plt.show()
```

```python
for band in [1, 2, 3]:
    wave, qe= load_qe(band=band)
    aa =  [(i.to(ur.AA)).value for i in wave]
    plt.scatter(aa, qe, label = 'Band {}'.format(band))
    plt.xlim([1300, 4000])
    plt.xlabel('Wavelength (Ang)')
    plt.ylabel('QE')


plt.legend()
plt.show()
```

```python
wave, reflectivity = load_reflectivity(diag=True)
aa =  [(i.to(ur.AA)).value for i in wave]
plt.scatter(aa, reflectivity)
plt.xlim([1300, 4000])
plt.ylim([80, 93])
plt.xlabel('Wavelength (Ang)')
plt.ylabel('Reflectivity')
plt.show()
```
